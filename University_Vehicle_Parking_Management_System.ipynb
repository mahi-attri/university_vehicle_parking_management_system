```python
import mysql.connector
import tkinter as tk
from tkinter import messagebox, simpledialog

class ParkingSystem:
    def __init__(self, root):
        self.root = root
        self.db = mysql.connector.connect(
            host="localhost",
            user="root",
            password="Attri@20589",
            database="parking_system"
        )
        self.cursor = self.db.cursor()
        self.username = None
        self.create_login_frame()

    def create_login_frame(self):
        self.login_frame = tk.Frame(self.root)
        self.login_frame.pack()

        tk.Label(self.login_frame, text="Username:").grid(row=0, column=0)
        self.username_entry = tk.Entry(self.login_frame)
        self.username_entry.grid(row=0, column=1)

        tk.Label(self.login_frame, text="Password:").grid(row=1, column=0)
        self.password_entry = tk.Entry(self.login_frame, show="*")
        self.password_entry.grid(row=1, column=1)

        login_button = tk.Button(self.login_frame, text="Login", command=self.login)
        login_button.grid(row=2, columnspan=2)
        login_button.config(bg="light green")  # Set background color to green

    def login(self):
        username = self.username_entry.get()
        password = self.password_entry.get()
        sql = "SELECT * FROM users WHERE username = %s AND password = %s"
        val = (username, password)
        self.cursor.execute(sql, val)
        user = self.cursor.fetchone()
        if user:
            messagebox.showinfo("Login", "Login successful!")
            self.username = username
            self.create_main_menu()
        else:
            messagebox.showerror("Login", "Invalid username or password. Please try again.")

    def create_main_menu(self):
        self.login_frame.destroy()

        self.main_menu_frame = tk.Frame(self.root)
        self.main_menu_frame.pack()

        tk.Label(self.main_menu_frame, text="Welcome to Parking Management System").pack()
        

        reserve_button = tk.Button(self.main_menu_frame, text="Reserve Parking Spot", command=self.reserve_parking_spot)
        reserve_button.pack()
        reserve_button.config(bg="LIGHT grey") 

        find_button = tk.Button(self.main_menu_frame, text="Find Carpool Partner", command=self.find_carpool_partner)
        find_button.pack()
        find_button.config(bg="light grey")  

        emergency_button = tk.Button(self.main_menu_frame, text="Request Emergency Assistance", command=self.request_emergency_assistance)
        emergency_button.pack()
        emergency_button.config(bg="light grey") 

        exit_button = tk.Button(self.main_menu_frame, text="Exit", command=self.exit_program)
        exit_button.pack()
        exit_button.config(bg="light red")  # Set background color to red

    def reserve_parking_spot(self):
        vehicle_type = simpledialog.askstring("Reserve Parking Spot", "Enter your vehicle type:")
        spot = simpledialog.askstring("Reserve Parking Spot", "Enter parking spot:")
        sql = "INSERT INTO reservations (username, vehicle_type, spot) VALUES (%s, %s, %s)"
        val = (self.username, vehicle_type, spot)
        self.cursor.execute(sql, val)
        self.db.commit()
        messagebox.showinfo("Reserve Parking Spot", "Parking spot reserved successfully!")

    def find_carpool_partner(self):
        destination = simpledialog.askstring("Find Carpool Partner", "Enter your destination:")
        sql = "SELECT username FROM carpool WHERE destination = %s"
        val = (destination,)
        self.cursor.execute(sql, val)
        carpools = self.cursor.fetchall()
        if carpools:
            partners = "\n".join(partner[0] for partner in carpools)
            messagebox.showinfo("Find Carpool Partner", f"Carpool partners available:\n{partners}")
        else:
            messagebox.showinfo("Find Carpool Partner", "No carpool partners available for your destination.")

    def request_emergency_assistance(self):
        problem = simpledialog.askstring("Request Emergency Assistance", "Describe your problem:")
        
        if problem:
            messagebox.showinfo("Request Emergency Assistance", "Emergency department has been notified of your problem. Help is on the way.")
        else:
            messagebox.showerror("Invalid Input", "Invalid input. Please enter a valid problem description.")

    def exit_program(self):
        self.root.destroy()
        self.cursor.close()

if __name__ == "__main__":
    root = tk.Tk()
    root.title("Parking Management System")
    app = ParkingSystem(root)
    root.mainloop()
```
